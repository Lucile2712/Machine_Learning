Q2: Train a decision tree for each of the previous configurations on the full dataset
Run a solid evaluation on different trees (with different parameters) by randomly splitting the data 80% for training and 20% for test *multiple times*.

```python
import csv
import numpy as np
from utils import load_from_csv
from sklearn import tree
from matplotlib import pyplot as plt # for a good visualization of the trees 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np

datas_examples, datas_labels, features, prediction = load_from_csv("./compass.csv")

trees_errors = []

# train test split 5 times
for i in range(5):
    # Split dataset into 80% training and 20% test set
    train_examples, test_examples, train_labels, test_labels = train_test_split(datas_examples, datas_labels, random_state=None, test_size=0.20)
    
    one_tree_errors = []

    trees = [
        DecisionTreeClassifier(splitter='random', max_depth=100, min_samples_leaf=10),
        DecisionTreeClassifier(splitter='random', max_depth=1, min_samples_leaf=10),
        DecisionTreeClassifier(splitter='random', max_depth=10, min_samples_leaf=50),
        DecisionTreeClassifier(splitter='best', max_depth=20, min_samples_leaf=10),
        DecisionTreeClassifier(splitter='best', max_depth=50, min_samples_leaf=10),
        DecisionTreeClassifier(splitter='best', max_depth=100, min_samples_leaf=50)
    ]

    # train and test every tree 
    for tree in trees:
        tree.fit(train_examples, train_labels)
        test_results = tree.predict(test_examples)
        error = np.sum(test_results != test_labels) / len(test_results)
        one_tree_errors.append(error)
    
    trees_errors.append(one_tree_errors)

for i, split_errors in enumerate(trees_errors):
    print(f"\n\nOne train-test separation errors:")
    for j, error in enumerate(split_errors):
        print(f"TREE{j+1}: {error}")
```

<div style="border: 1px solid green; padding: 10px; margin-top: 10px; border-radius: 5px">
Answer :

We changed the training and testing examples every time for each tree. We notice the rate of errors varies but stays around the 33% range.
We notice that using the "best" splitter we have consistently better results than using the random one.
When the max depth is 1, we get a lot of mistakes. It makes sense since the tree is severely underfit.

</div>


Q3: Propose an evaluation in terms of training and testing accuracies using $5$-cross validation on two decision trees that have different topologies

Q4: Propose an experimental study that shows the transition phase from underfitting to overfitting 
Evaluate the impact (in terms of accuracy) of the three parameters : maximum depth, splitting criterion, and the minimum sample leafs. 

```python

train_examples, train_labels, features, prediction = load_from_csv("./compass.csv")


#Tests to visualise underfitting and overfitting

for i in range(1,10):
    clf = tree.DecisionTreeClassifier(splitter='best', max_depth=i*2)

    scores = cross_val_score(clf, datas_examples, datas_labels, cv=5)
    print('scores for max depth:', i*2, " is ", scores)
    mean_accuracy = scores.mean()
    print('Mean accuracy: %.2f' % mean_accuracy)
    clf.fit(train_examples, train_labels)
    prediction = clf.predict(test_examples)
    #print the error of the prediction
    nb_error = np.sum(prediction != test_labels)/len(prediction)
    print('error: ', nb_error)

for i in range(1,25):
    clf = tree.DecisionTreeClassifier(splitter='best', min_samples_leaf=i*10)

    scores = cross_val_score(clf, datas_examples, datas_labels, cv=5)
    print('scores for min samples leaf:', i*10, " is ", scores)
    mean_accuracy = scores.mean()
    print('Mean accuracy: %.2f' % mean_accuracy)
    clf.fit(train_examples, train_labels)
    prediction = clf.predict(test_examples)
    #print the error of the prediction
    nb_error = np.sum(prediction != test_labels)/len(prediction)
    print('error: ', nb_error)
```
